```python
import yfinance as yf
import pandas as pd

# Danh sách ticker các công ty
tickers = ["AAPL", "MSFT", "GOOGL", "AMZN", "TSLA", "META", "NVDA", "F", "JNJ", "JPM",
           "V", "PG", "UNH", "XOM", "NFLX"]

# Dữ liệu sẽ được lưu trữ trong danh sách
financial_data = []

# Lặp qua các mã cổ phiếu trong danh sách
for ticker in tickers:
    # Lấy dữ liệu từ Yahoo Finance
    stock = yf.Ticker(ticker)
    
    # Lấy báo cáo tài chính theo năm
    income_stmt = stock.financials
    balance_sheet = stock.balance_sheet
    cashflow_stmt = stock.cashflow
    key_stats = stock.info
    
    # Kiểm tra xem có dữ liệu hay không
    if income_stmt.empty or balance_sheet.empty:
        print(f"Không có dữ liệu tài chính cho {ticker}")
        continue

    # Lặp qua các năm từ 2019 đến 2024
    for year in range(2019, 2025):
        # Kiểm tra năm có dữ liệu hay không
        if year not in income_stmt.columns or year not in balance_sheet.columns:
            print(f"Không có dữ liệu cho {ticker} vào năm {year}")
            continue
        
        # Lấy các giá trị từ các báo cáo tài chính
        net_income = income_stmt.loc['Net Income', year] if 'Net Income' in income_stmt.index else None
        total_revenue = income_stmt.loc['Total Revenue', year] if 'Total Revenue' in income_stmt.index else None
        total_assets = balance_sheet.loc['Total Assets', year] if 'Total Assets' in balance_sheet.index else None
        total_equity = balance_sheet.loc['Total Stockholder Equity', year] if 'Total Stockholder Equity' in balance_sheet.index else None
        total_debt = balance_sheet.loc['Total Debt', year] if 'Total Debt' in balance_sheet.index else None
        eps = income_stmt.loc['Earnings Per Share', year] if 'Earnings Per Share' in income_stmt.index else None
        dividend_yield = key_stats.get("dividendYield", None)  # Dividend yield từ thông tin chung
        
        # Tính toán các chỉ số tài chính
        roa = net_income / total_assets if total_assets and net_income else None
        roe = net_income / total_equity if total_equity and net_income else None
        debt_equity = total_debt / total_equity if total_debt and total_equity else None
        pe_ratio = key_stats.get('trailingPE', None)  # P/E ratio
        revenue = total_revenue
        net_income_value = net_income

        # Thêm dữ liệu vào danh sách
        financial_data.append({
            "Date": year,
            "Ticker": ticker,
            "ROA": roa,
            "ROE": roe,
            "EPS": eps,
            "P/E Ratio": pe_ratio,
            "Debt/Equity": debt_equity,
            "Dividend Yield": dividend_yield,
            "Revenue": revenue,
            "Net Income": net_income_value
        })

# Chuyển dữ liệu sang DataFrame
df_financials = pd.DataFrame(financial_data)

# Lưu dữ liệu vào file CSV
df_financials.to_csv("yahoo_financial_data_2019_2024.csv", index=False)

print("Đã tải dữ liệu tài chính từ 2019-2024 thành công!")
```

```python
import yfinance as yf
import pandas as pd
import numpy as np

# Danh sách ticker
tickers = ["AAPL", "MSFT", "GOOGL", "AMZN", "TSLA", "META", "NVDA", "F", "JNJ", "JPM",
           "V", "PG", "UNH", "XOM", "NFLX"]

financial_data = []

# Lặp qua từng ticker
for ticker in tickers:
    stock = yf.Ticker(ticker)
    info = stock.info

    # Lấy các trường theo đúng yêu cầu (nếu không có thì dùng np.nan)
    roa = info.get("returnOnAssets", np.nan)
    roe = info.get("returnOnEquity", np.nan)
    eps = info.get("trailingEps", np.nan)
    pe_ratio = info.get("trailingPE", np.nan)
    debt_equity = info.get("debtToEquity", np.nan)
    dividend_yield = info.get("dividendYield", np.nan)
    revenue = info.get("totalRevenue", np.nan)
    net_income = info.get("netIncomeToCommon", np.nan)

    # Lặp qua các năm từ 2019 đến 2024
    for year in range(2019, 2025):
        # Gán ngày cuối năm để đảm bảo khoảng thời gian từ 01/01/2019 đến 31/12/2024
        date_str = f"{year}-12-31"
        financial_data.append({
            "Date": date_str,
            "Ticker": ticker,
            "ROA": roa,
            "ROE": roe,
            "EPS": eps,
            "P/E Ratio": pe_ratio,
            "Debt/Equity": debt_equity,
            "Dividend Yield": dividend_yield,
            "Revenue": revenue,
            "Net Income": net_income
        })

# Chuyển dữ liệu thành DataFrame
df_financials = pd.DataFrame(financial_data)

# Xuất dữ liệu ra file CSV
df_financials.to_csv("yahoo_financial_data.csv", index=False)

print("✅ Đã tải dữ liệu tài chính thành công!")
```

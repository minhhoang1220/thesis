```python
import yfinance as yf
import pandas as pd

# Danh sách ticker (cập nhật theo danh sách doanh nghiệp bạn cần)
tickers = ["AAPL", "MSFT", "GOOGL", "AMZN", "TSLA", "META", "NVDA", "F", "JNJ", "JPM",
           "V", "PG", "UNH", "XOM", "NFLX"]  # Thêm mã VN nếu có trên Yahoo Finance

# Lấy dữ liệu lịch sử trong khoảng 5 năm (2020 - 2024)
start_date = "2019-01-01"
end_date = "2024-12-31"

data_list = []

for ticker in tickers:
    try:
        stock_data = yf.download(ticker, start=start_date, end=end_date)
        if stock_data.empty:
            print(f"Không có dữ liệu cho mã {ticker}.")
            continue
        stock_data["Ticker"] = ticker
        data_list.append(stock_data)
    except Exception as e:
        print(f"Lỗi khi tải dữ liệu cho mã {ticker}: {e}")

# Gộp dữ liệu
if data_list:
    df_price = pd.concat(data_list)
    df_price.reset_index(inplace=True)

    cols_needed = ["Date", "Ticker", "Open", "High", "Low", "Close", "Volume"]
    if "Adj Close" in df_price.columns:
        cols_needed.append("Adj Close")
    df_price = df_price[cols_needed]


    # Xuất file CSV
    output_file = "yahoo_price_data2.csv"
    df_price.to_csv(output_file, index=False)
    print(f"Đã tải dữ liệu giá cổ phiếu thành công! File được lưu tại: {output_file}")
else:
    print("Không có dữ liệu để lưu.")
```

```python
import pandas as pd

# Đọc dữ liệu từ CSV
csv_file = "yahoo_price_data2.csv"  # Thay bằng file bạn muốn chuyển đổi
df = pd.read_csv(csv_file)

# Lưu dưới dạng Excel
excel_file = "yahoo_price_data.xlsx"
df.to_excel(excel_file, index=False, engine="openpyxl")

print(f"Đã chuyển đổi {csv_file} thành {excel_file} thành công!")
```
